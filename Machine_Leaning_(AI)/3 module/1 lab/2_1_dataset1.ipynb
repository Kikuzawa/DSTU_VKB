Визуализация распределения возраста

```python
features.replace('undefined', np.nan, inplace=True)
#визуализация распределения возраста
sns.set(color_codes=True)
plot = sns.distplot(features.Age.dropna())
plot.figure.set_size_inches(6,6) # в дюймах
```

```python
sns.boxplot(x=features["Age"].dropna())
```

```python
features.loc[(features["Age"] < 14) | (features["Age"] > 100), "Age"] = np.nan
print(features.isnull().sum()["Age"])
```

```python
%matplotlib inline   
import seaborn as sns 
sns.set_theme(color_codes=True)
plot = sns.displot(features["Age"].dropna())
plot.figure.set_size_inches(6,6) # в дюймах
```

```python
#обработка nan значений, замена их на среднее значение по столбцу
features["Age"] = features["Age"].fillna(features["Age"].mean())
print(features.isnull().sum()["Age"])
print(features["Age"].max())
sns.boxplot(x=features["Age"].dropna())
```

10. Поиск неявных дубликатов может быть проведен так

```python
def clean(response): 
    if isinstance(response, str):
        return response[0].upper() + response[1:].lower()
    return response

features["id"] = features["id"].apply(lambda x: clean(x)) 
print(features.head())
```

1. Создать группы по нескольким столбцам датасета

```python
dt = pd.read_csv("data.csv")
group = dt.groupby(["City"])
```

2. Получить информацию о количестве групп, которые будут созданы

```python
group.ngroups
```

3. Получить информацию о количестве элементов в каждой группе

```python
group.size()
```

4. Просмотр полученных групп

```python
group.groups
```

5. Просмотр данных конкретной группы

```python
group.get_group(('Faridabad',))
```

6. Обходы групп: вычисление среднего

```python
group["Age"].mean()
```

7. Различные типы группировок: по столбцам, по уровням индекса

```python
# По столбцам
group = dt.groupby(['Age', 'City'])
group.size()
```

```python
# По уровням индекса
copy_dt = dt.copy()
copy_dt = copy_dt.set_index(['Age', 'City'])
group = copy_dt.groupby(level=['Age'])
group.size()
```

8. Группировка по одному или нескольким уровням индекса

```python
# Один уровень индекса
copy_dt = dt.copy()
copy_dt = copy_dt.set_index(['Age', 'City'])
group = copy_dt.groupby(level=['Age'])
group.size()
```

9. Агрегирование групп

```python
group_type_1 = dt.groupby(['Study Satisfaction'])['Academic Pressure']

group_type_1.max()
```

10. Преобразование групп: метод transform, метод apply

```python
dt.groupby('City').transform('sum')
```

11. Произвести исключение групп

```python
f = lambda x: x['Age'].count() > 1
grouped_data = dt.groupby(['Age', 'City']).filter(f)
grouped_data
```
