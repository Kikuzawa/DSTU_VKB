СПИСКИ

Списки - это изменяемые последовательности, которые похожи на массивы. Списки могут одновременно хранить в себе значения разных типов. Отличие списка от массива состоит в том, что элементы списка непоследовательно расположены в памяти, поэтому доступ к ним более медленный в сравнении с той же операцией для массива.

```python
a = [1, 10,100] # Список с элементами 1, 10, 100
b = [1, 2, [3, 4], 5] # Список [3, 4] вложен в список b

print()
print(f'максимальное значение в списке а = {max(a)}')
print(f'минимальное значение в списке а = {min(a)}')

a.append(5)  #добавление элемента в список
print(f'список a с новым элементом = {a}')

a.insert(0, 3) #текущее значение элемента с индексом 0 заменить на значение равное 3
print(f'список a с новым элементом = {a}')

a.extend([3, 4]) # добавление нескольких значений в список
print(f'список a с новым элементом = {a}')
```

Доступ к элементам списка осуществляется с помощью индексации , которая начинается с 0, и оператора [ ].

```python
a = [4, 3, 2, 1, 2]
print(a[0]) # обращение к элементу с индексом 0
print(a[-1]) # индексация с конца списка
print(a[2:4]) # срез списка от 2-ого эл-то до 4-ого по индексам
print(a[:2])
print(a[2:])
print(a[:])
print()
a.remove(2) #поиск и удаление элемента по значению
print(f'после удаления первой 2 = {a}')
a.pop(1) # или a.pop()
print(f'после удаления элемента с индексом 1 = {a}')
```

оператор IF возвращает TRUE, когда выражение-условие не равно нулю, не является пустым, является логическим.

```python
a = 0
b = 34
c = 10
if c < 100:
    a = b + c
print(a)
```

оператор ELSE является необязательным в связке с IF, задает альтернативный ход вычислений.

```python
a = 7
if a > 5:
    print("Да")
else:
    print("Нет")
```

Оператор elif и конструкция if – elif – else

Оператор elif позволяет упростить код. Сделать его легче читаемым. Позволяет избежать написание несколько условий if в Python.

```python
balance = 500
if balance < 0:
    print("Баланс ниже нуля, положите деньги на счет, чтобы Вам не выписали штраф")
elif balance == 0:
    print("Баланс равен нулю, скорее внесите деньги на счет")
else:
    print("Ваш баланс выше нуля, все хорошо")
```

```python
#обработка исключений
inp = input('введите температуру по Фаренгейту:')
print("Вы ввели :", inp)
try:
    fahr = float(inp)
    cel = (fahr - 32.0) * 5.0 / 9.0
    print(cel)
except:
    print('Пожалуйста, введите число')
```

После тела цикла можно указать else и блок операций, которые необходимо выполнить, когда закончится while.

```python
count = 3
while count < 7:
    print (count, " меньше 7")
    count = count + 1
else:
    print (count, " не меньше 7")
```

Оператор break используется для выхода из цикла Python — прерывает его досрочно.

```python
while True:
    name = input("Введите имя:")
    print("Привет", name)
    if name == "хватит":
        break
print("алгоритм дошел до break")
```

ФУНКЦИИ

Функция — это фрагмент программного кода, который решает какую-либо задачу.

image.png

```python
# функция из примера ниже принимает два множителя, а возвращает их округленное
# до целого числа произведение
def int_multiple(a, b):
    product = a * b
    # возвращаем значение
    return int(product)


print(int_multiple(341, 2.7))
```

```python
# пузырьковая сортировка
def bubble_sort(nums):
    for i in range(0, len(nums) - 1):
        for j in range(len(nums) - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
    return nums

print(bubble_sort([2,4,6,11,90,1,0]))
```
