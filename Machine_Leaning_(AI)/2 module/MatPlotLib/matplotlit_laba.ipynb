# 2.Линейная диаграмма

```python
import matplotlib.pyplot as plt

# Построение горизонтальной линии на уровне y=1
plt.axhline(y=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Построение горизонтальных линий разных типов, ширины и цветов
plt.axhline(y=2, color='r', linestyle='-', linewidth=2)  # Красная штрих линия
plt.axhline(y=3, color='g', linestyle=':', linewidth=1)   # Зеленая пунктирная линия
plt.axhline(y=4, color='b', linestyle='--', linewidth=3)   # Синяя сплошная линия
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Создание данных для графика
x = np.linspace(0, 20, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Построение графика и заполнение между двумя горизонтальными линиями
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)  # Горизонтальная линия на уровне y=0
plt.fill_between(x, y1, y2, where=(y1 > y2), alpha=0.5, color='green')
plt.legend()
plt.show()
```

# 3.Гистограмма

```python
import matplotlib.pyplot as plt

# Данные для гистограммы
data = [5, 20, 30, 45, 50]

# Создание горизонтальной гистограммы
plt.barh(range(len(data)), data)

# Отображение графика
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные для гистограммы
data = [5, 20, 30, 45, 50]

# Создание горизонтальной гистограммы
plt.barh(range(len(data)), data)

# Аннотирование значений столбцов
for i, value in enumerate(data):
    plt.text(value, i, str(value))

# Отображение графика
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные для гистограммы
data = [5, 20, 30, 45, 50]

# Создание горизонтальной гистограммы
plt.barh(range(len(data)), data)

# Аннотирование столбцов
for i, value in enumerate(data):
    plt.annotate(str(value), xy=(value, i))

# Отображение графика
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные для сгруппированного столбчатого графика
categories = ['Категория А', 'Категория Б', 'Категория В']
values1 = [10, 20, 30]
values2 = [15, 25, 35]

# Создание сгруппированного столбчатого графика
fig, ax = plt.subplots()

bar_width = 0.35
index = range(len(categories))

bar1 = ax.bar(index, values1, bar_width, label='Group 1')
bar2 = ax.bar([i + bar_width for i in index], values2, bar_width, label='Group 2')

# Аннотирование столбцов
for i, v in enumerate(values1):
    ax.text(i, v + 2, str(v), ha='center')

for i, v in enumerate(values2):
    ax.text(i + bar_width, v + 2, str(v), ha='center')

# Добавление подписей на оси X
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(categories)

# Отображение графика
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV файла
dataset = pd.read_csv('housing_market_dataset.csv')

# Выбор столбца "Класс жилья" и подсчёт частоты значений
type_counts = dataset['Класс жилья'].value_counts()

# Сортировка данных по убыванию
type_counts = type_counts.sort_values(ascending=False)

# Изображение столбчатой диаграммы
plt.bar(type_counts.index, type_counts.values)
plt.xticks(rotation=30, ha='right')
plt.xlabel('Класс жилья')
plt.ylabel('Число объявлений')
plt.title('Распределение объявлений по классу жилья')
plt.show()

# Вывод порядка убывания размеров столбцов
print('Порядок убывания размеров столбцов:', ''.join(map(str, type_counts.index)))
```
